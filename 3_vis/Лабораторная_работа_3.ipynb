```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/data.csv")
# df = pd.read_csv("./data/update.csv")
```

Общая картина по основным предметам

```python
plt.figure(figsize = (16, 9))
sns.set(font_scale = 1.35)
dataset = pd.DataFrame(data = df, columns = ["math", "literature", 'english', "physics", "chemistry", "biology", "history", "geography", "civic_education"])
boxplot_dataset = sns.boxplot(x="variable", y="value", data=pd.melt(dataset), showfliers = False)
boxplot_dataset.set_xlabel("")
boxplot_dataset.set_ylabel("Оценки")
plt.title("Распределение оценок по основным предметам")
plt.show()
```

Иностранные языки: оценки и выбор

```python
plt.figure(figsize = (16, 9))
sns.set(font_scale = 1.35)
dataset = pd.DataFrame(data = df, columns = ["english", "german", "chinese", "russian", "japanese", "french"])
boxplot_dataset = sns.boxplot(x="variable", y="value", data=pd.melt(dataset), showfliers = False)
boxplot_dataset.set_xlabel("")
boxplot_dataset.set_ylabel("Оценки")
plt.title("Оценки по иностранным языкам")
plt.show()
```

```python
num_eng = sum(df.english >= 0)
num_russ = sum(df.russian >= 0)
num_french = sum(df.french >= 0)
num_chinese = sum(df.chinese >= 0)
num_german = sum(df.german >= 0)
num_japan = sum(df.japanese >= 0)
num_null = sum(df.id_examinee != None) - num_eng - num_russ - num_french - num_chinese - num_german - num_japan


language = ["English", "Russian", "French", "Chinese", "German", "Japanese"]
choice = [int(num_eng), int(num_russ), int(num_french), int(num_chinese), int(num_german), int(num_japan)]
fig, ax = plt.subplots(figsize=(12, 6))
bars = plt.barh(language, choice, color = "wheat")


for  bar in bars:
    width = bar.get_width()
    label_y = bar.get_y() + bar.get_height()/2
    plt.text(width, label_y, s = f"{width}", color = "r")
    
plt.xlabel("Количество студентов")
plt.ylabel("")
plt.title("Выбор иностранного языка")    
plt.show
```

Выбор в пользу соц или гуманитарных наук:

```python
sum_of_NS = sum((df.physics >= 0) | (df.chemistry >= 0) | (df.biology >= 0))
sum_of_SS = sum((df.history >= 0) | (df.geography >= 0) | (df.civic_education >= 0))
sum_of_Null = sum(df.id_examinee >= 0) - sum_of_NS - sum_of_SS

# Bar chart for combination exam selection
labels = ["Social Science", "Natural Science", "?"]
count = [sum_of_SS, sum_of_NS, sum_of_Null]
colors = ["wheat", "plum" ,"grey"]

plt.figure(figsize = (5, 5))
plt.pie(count, labels = labels, colors = colors, autopct='%0.2f%%')
plt.title("Частота выбора комбинированного экзамена")
plt.show()
```

```python
math_papers = df[["math"]]
math_dataset_cleaned = math_papers.dropna()

physics_papers = df[["physics"]]
physics_dataset_cleaned = physics_papers.dropna()

english_papers = df[["english"]]
english_dataset_cleaned = english_papers.dropna()

chemistry_papers = df[["chemistry"]]
chemistry_dataset_cleaned = chemistry_papers.dropna()

literature_papers = df[["literature"]]
literature_dataset_cleaned = literature_papers.dropna()

biology_papers = df[["biology"]]
biology_dataset_cleaned = biology_papers.dropna()

history_papers = df[["history"]]
history_dataset_cleaned = history_papers.dropna()

geography_papers = df[["geography"]]
geography_dataset_cleaned = geography_papers.dropna()

civic_education_papers = df[["civic_education"]]
civic_education_dataset_cleaned = civic_education_papers.dropna()

math_average = np.average(math_dataset_cleaned)
physics_average = np.average(physics_dataset_cleaned)
chemistry_average = np.average(chemistry_dataset_cleaned)
english_average = np.average(english_dataset_cleaned)
literature_average = np.average(literature_dataset_cleaned)
biology_average = np.average(biology_dataset_cleaned)
history_average = np.average(history_dataset_cleaned)
geography_average = np.average(geography_dataset_cleaned)
civic_education_average = np.average(civic_education_dataset_cleaned)

math_round_score = round(math_average, 2)
physics_round_score =  round(physics_average, 2)
chemistry_round_score = round(chemistry_average, 2)
biology_round_score = round(biology_average, 2)
english_round_score = round(english_average, 2)
literature_round_score = round(literature_average, 2)
history_round_score = round(history_average, 2)
geography_round_score = round(geography_average, 2)
civic_education_round_score = round(civic_education_average, 2)

labels_average = ["Math", "Literature", "English", "Physics", "Chemistry", "Biology", "History",
                    "Geography", "Civic Education"]
counts_average_round = [math_round_score, literature_round_score, english_round_score, physics_round_score, chemistry_round_score,
                        biology_round_score, history_round_score, geography_round_score, civic_education_round_score]

fig, ax = plt.subplots(figsize = (16, 9))
sns.set(font_scale = 1.35)
score_average = sns.barplot({"x": labels_average, "y": counts_average_round}, x='x', y='y')

for counts_10 in score_average.patches:
  height = counts_10.get_height()
  label_x = counts_10.get_x() + counts_10.get_width() / 2
  label_y = counts_10.get_y() / height + height
  plt.text(label_x, label_y, s = f"{height:}", ha='center', va='bottom', color = "black", size = 15)

plt.xlabel("")
plt.ylabel("Оценки")
plt.title("Средний балл")
```

```python
math_10_score = sum(df.math == 10)
literature_10_score = sum(df.literature == 10)
english_10_score = sum(df.english == 10)
physics_10_score = sum(df.physics == 10)
chemistry_10_score = sum(df.chemistry == 10)
biology_10_score = sum(df.biology == 10)
history_10_score = sum(df.history == 10)
geography_10_score = sum(df.geography == 10)
civic_education_10_score = sum(df.civic_education == 10)

labels_10_score = ["Math", "Literature", "English", "Physics", "Chemistry", "Biology", "History",
                    "Geography", "Civic Education"]
counts_10_score = [math_10_score, literature_10_score, english_10_score, physics_10_score, chemistry_10_score,
                    biology_10_score, history_10_score, geography_10_score, civic_education_10_score]

fig, ax = plt.subplots(figsize = (16, 9))
sns.set(font_scale = 1.35)
score_10_grouped = sns.barplot({"x":labels_10_score, "y": counts_10_score}, x='x', y='y')
score_10_grouped.set_xticklabels(score_10_grouped.get_xticklabels())
ax.bar_label(ax.containers[0])

plt.xlabel("")
plt.ylabel("Полученные оценки")
plt.title("Частота получения максимального балла")
```

```python
math_failed = sum(df.math <= 1.0)
literature_failed = sum(df.literature <= 1.0)
english_failed = sum(df.english <= 1.0)
physics_failed = sum(df.physics <= 1.0)
chemistry_failed = sum(df.chemistry <= 1.0)
biology_failed = sum(df.biology <= 1.0)
history_failed = sum(df.history <= 1.0)
geography_failed = sum(df.geography <= 1.0)
civic_education_failed = sum(df.civic_education <= 1.0)

list_failed = ["Math", "Literature", "English", "Physics", "Chemistry", "Biology",
              "History", "Geography", "Civic Education"]

counts_failed = [math_failed, literature_failed, english_failed, physics_failed, chemistry_failed,
                 biology_failed, history_failed, geography_failed, civic_education_failed]

plt.figure(figsize = (16, 9))
plt.pie(counts_failed, labels = list_failed, autopct='%0.2f%%')
plt.title("Частота получения непроходного балла")
plt.show()
```

При выборе языка доминирует английский, хотя оценки по нему в среднем ниже, чем для остальных языков. Студенты чаще предпочитают соц науки естественным. Инфляция оценок по "гражданскому воспитанию" - это может объяснять популярность соц наук. Сложность экзамена по истории
